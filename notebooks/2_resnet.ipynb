### Running and Viewing Results

To run and view the results of the ResNet models:

- **ResNet Base Model**  
  Navigate to the `resnet_base` folder. This directory contains the notebook with all training results and evaluation outputs for the baseline model.

- **ResNet with Phase Training and Class Weights**  
  Navigate to the `resnet_with_phase_training_with_classweights` folder. This directory contains the notebook demonstrating the phased training approach along with class weighting, including results and performance metrics.

Each folder includes a self-contained notebook that can be executed to reproduce the results.
